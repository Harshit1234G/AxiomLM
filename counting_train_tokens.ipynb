```python
import sys

sys.path.append('kaggle/input/axiom-utils')
import llm_components as lc
```

```python
sp = lc.load_sp_tokenizer()

total_tokens = 0
with open('kaggle/input/wikitext-103/wikitext-103/wiki.train.tokens', 'r', encoding='utf-8') as f:
    for line in f:
        ids = sp.encode(line, out_type= int)
        total_tokens += len(ids)

print(f'Total tokens: {total_tokens:3,}')
```

```python
steps_per_epoch = total_tokens // (lc.BATCH_SIZE * lc.SEQUENCE_LEN)
print(f'Steps Per Epoch: {steps_per_epoch:3,}')
```

```python
num_epochs = 2
total_steps = steps_per_epoch * num_epochs
print(f'Total Steps: {total_steps:3,}')
```

```python
print(f'Warmup Steps: {int(total_steps * 0.05)}')
```
