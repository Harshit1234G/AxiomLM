```python
import sys

sys.path.append('kaggle/input/axiom-utils')
import llm_components as lc
```

```python
sp = lc.load_sp_tokenizer()
loader = lc.LMDatasetLoader(sp)
ds = loader.create('kaggle/input/wikitext-103/wikitext-103/wiki.train.tokens', training= False)
```

```python
steps_per_epoch = 0
for _ in iter(ds):
    steps_per_epoch += 1
```
