```python
import sentencepiece as spm
```

```python
train_path = 'kaggle/input/wikitext-103/wikitext-103/wiki.train.tokens'
out_path = 'kaggle/input/axiom-utils/sp_tokenizer'
```

```python
spm.set_random_generator_seed(42)
```

```python
spm.SentencePieceTrainer.train(
    input= train_path,
    model_prefix= out_path,
    vocab_size= 16000,
    model_type= 'bpe',
    bos_id= 1,
    eos_id= 2,
    unk_id= 0,
    pad_id= 3,
    character_coverage= 0.9995,
    normalization_rule_name= 'identity',
    input_sentence_size= 50_000_000,
    shuffle_input_sentence= True,
    num_threads= 4
)
```

```python
sp = spm.SentencePieceProcessor()
sp.load(out_path + '.model')
```

```python
sp.encode(['This is a test', 'Hello world'], out_type= int)
```

```python
sp.encode(['This is a test', 'Hello world'], out_type= str)
```

```python
sp.decode([[608, 120, 6, 1837], [6740, 15920, 1036]])
```

```python
sp.decode([['▁This', '▁is', '▁a', '▁test'], ['▁Hell', 'o', '▁world']])
```
